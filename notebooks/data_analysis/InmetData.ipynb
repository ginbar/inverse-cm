```python
import pandas as pd
import os
import glob
```

```python
folder_path = "../../data/inmet-am"
inmet_files = glob.glob(os.path.join(folder_path, "*.CSV"))
n_files = len([name for name in os.listdir(folder_path)])
```

```python
dfs = []

for f in inmet_files[1:]:
    df = pd.read_csv(f, sep=';', skiprows=8, encoding="iso-8859-1")
    dfs.append(df.groupby("Data")["UMIDADE RELATIVA DO AR, HORARIA (%)"].mean())

humidity_df = pd.concat(dfs, axis=1)

daily_mean_humidity = humidity_df.mean(axis=1)

daily_mean_humidity
    
```

```python
daily_mean_humidity.to_csv("../../data/daily_mean_humidity.csv")
```
